```python
# import dependencies 
import os
import pandas as pd
import time
import matplotlib.pyplot as plt
```

```python
# generate path to csv
csv_path = os.path.join('C:/Web Design Homework/Web-Design-Challenge/WebVisualizations/Resources/cities.csv')
cities_csv = pd.read_csv(csv_path)
cities_csv.head()
```

```python
cities_csv = cities_csv.rename(columns={'Max Temp': 'Max_Temp', 'Wind Speed': 'Wind_Speed'})
cities_csv.head()
```

```python
# create html table and save to file 
html_file = cities_csv.to_html('table.html', index=False, classes=['table', 'table-striped', 'table-hover'])
```

```python
# Build the scatter plots for latitude vs. cloudiness

x_axis = cities_csv.Lat
y_axis = cities_csv.Max_Temp

plt.scatter(x_axis, 
            y_axis,
            edgecolor="black", color = "c", linewidths=1, marker="o", 
            alpha=0.75, label="City")

# Incorporate the other graph properties
plt.title("City Latitude vs. Max Temperature (F) (%s)" % time.strftime("%x"))
plt.ylabel("Max Temp (F)")
plt.xlabel("Latitude")
plt.grid(False)

# Save the figure
plt.savefig("C:/Web Design Homework/Web-Design-Challenge/WebVisualizations/img1.png")

# Show plot
plt.show()
```
